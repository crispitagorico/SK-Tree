## Example: Brownian tree

```python
tree_depth = 4
path_length, path_dim = 100, 3
min_branches, max_branches = 2, 5
min_steps, max_steps = 100, 200
```

```python
root_path = AddTime().fit_transform([brownian(path_length, path_dim-1)])[0]
value = (1.,root_path)
```

```python
tree = generate_brownian_tree(tree_depth, path_dim-1, value, min_branches, max_branches, min_steps, max_steps, perturb=False)
tree_ = generate_brownian_tree(tree_depth, path_dim-1, value, min_branches, max_branches, min_steps, max_steps, perturb=True)
```

# Mahalanobis distance for anomaly detection

```python
# N is the number of standard trees; N_perturbed is the number of anomalous trees
N, N_perturbed = 100, 10

# generate dataset of standard trees
list_of_trees = [generate_brownian_tree(tree_depth, path_dim-1, value, min_branches, max_branches, 
                                        min_steps, max_steps, rand=False, perturb=False) for k in range(N)]

# add some anomalies at the end
list_of_trees = list_of_trees + [generate_brownian_tree(tree_depth, path_dim-1, value, min_branches, max_branches, min_steps, 
                                                        max_steps, rand=False, perturb=True) for k in range(N_perturbed)]
```

```python
cd = treeMahalanobisDistance(list_of_trees,list_of_trees, 3)
```

```python
sns.heatmap(cd, cmap='Blues')
plt.show()
```
